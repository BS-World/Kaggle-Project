# <h1 style="background-color:#FF671F; font-weight:bold; text-align:center; color:white;" >ALIRAJPUR</h1>

```python
#No. of eligible families
Df_eligible_families_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['No. of eligible families'].sum()
print('Number of eligible families in ALIRAJPUR:-', Df_eligible_families_ALIRAJPUR)
```

```python
#Eligible families with Aadhar
Df_eligible_families_Aadhar_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['Eligible families with Aadhar'].sum()
print('Number of Eligible families with Aadhar in ALIRAJPUR:-', Df_eligible_families_Aadhar_ALIRAJPUR)
```

```python
Df_eligible_families_without_Aadhar_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['Eligible families without Aadhar'].sum()
print('Number of Eligible families without Aadhar in ALIRAJPUR:-', Df_eligible_families_without_Aadhar_ALIRAJPUR)
```

```python
#No. of eligible population
Df_eligible_population_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['No. of eligible population'].sum()
print('Number of Eligible population in ALIRAJPUR:-', Df_eligible_population_ALIRAJPUR)
```

```python
#Eligible population with Aadhar
Df_eligible_population_with_Aadhar_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['Eligible population with Aadhar'].sum()
print('Number of Eligible population with Aadhar in ALIRAJPUR:-', Df_eligible_population_with_Aadhar_ALIRAJPUR)
```

```python
#Eligible population without Aadhar
Df_eligible_population_without_Aadhar_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['Eligible population without Aadhar'].sum()
print('Number of Eligible population without Aadhar in ALIRAJPUR:-', Df_eligible_population_without_Aadhar_ALIRAJPUR)
```

```python
#Aadhar Enrollments in State as on 29-05-15
Df_Enrollments_Aadhar_ALIRAJPUR = df[df['District'] == 'ALIRAJPUR']['Aadhar Enrollments in State as on 29-05-15'].sum()
print('Number of Enrollments Aadhar in ALIRAJPUR:-', Df_Enrollments_Aadhar_ALIRAJPUR)
```

```python
district_data = df[df['District'] == 'ALIRAJPUR']

# Extract the values for the pie chart
labels = ['Eligible families with Aadhar', 'Eligible families without Aadhar', 'Eligible population with Aadhar', 'Eligible population without Aadhar']
values = district_data.loc[:, ['Eligible families with Aadhar', 'Eligible families without Aadhar', 'Eligible population with Aadhar', 'Eligible population without Aadhar']].values.flatten()

# Create the pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('ALIRAJPUR District - Aadhar Seeding Status')
plt.axis('equal')
plt.show()
```

```python
filtered_df = df[df['District'] == 'ALIRAJPUR']

# Select the desired columns for visualization
columns = [
    'No. of eligible families',
    'No. of eligible population',
    'Eligible families with Aadhar',
    'Eligible families without Aadhar',
    'Eligible population with Aadhar',
    'Eligible population without Aadhar'
]

# Create a bar chart for the selected columns
filtered_df[columns].plot(kind='bar', figsize=(10, 6))
plt.title("Data Visualization for ALIRAJPUR District")
plt.xlabel("Columns")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```
